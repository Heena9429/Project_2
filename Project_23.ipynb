# Data Visulaization

#### Importing Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Line plot
x = np.linspace(0.2, 0.46, 100)
y = np.sin(x)

# Create a new figure
plt.figure(figsize=(10,6))

# Plot the data
plt.plot(x,y)

# Add lables and title
plt.xlabel('Temprature')
plt.ylabel('Season')
plt.title('Line Plot')

# Show the plot
plt.show()
```

```python
# Bar Graph
season = ['winter', 'Summer', 'Fall', 'Spring']
weather = ['few clouds', 'mist', 'light rain', '4']

# Create the plot
plt.figure(figsize=(10, 6))
bars = plt.bar(season, weather, color=sns.color_palette("viridis", 5))

# Customize the plot
plt.title('Bar Graph', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Weather', fontsize=12)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

plt.show()
```

# Outliers

#### Importing Libraries

```python
import scipy
```

```python
numeric_cols=final_df.select_dtypes(include=['number']).columns
Q1=final_df[numeric_cols].quantile(0.25)
Q3=final_df[numeric_cols].quantile(0.75)
IQR = Q3-Q1
```

```python
outliers = ((final_df[numeric_cols] < (Q1 - 1.5 * IQR)) | (final_df[numeric_cols] > (Q3 + 1.5 * IQR))).sum()
print("Outliers  Count Per Column:\n", outliers)
```
